```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats
import time
```

### Heston Model

```python
# Simulation Setting
T = 1.0             # Time
N = 252             # The number of split
dt = T/N            # Delta time
sim = 100           # The number of simulations

# Initial Value
S0 = 100.0          # Initial Stock Price
v0 = 0.5            # Initial Volatility

# Constants of Heston Process
ksi = 0.5           # Volatility of Volatility
kappa = 0.3         # Kappa; the rate at which νt reverts to θ.
theta = 0.5**2      # Theta;  the long variance, or long-run average variance of the price; as t tends to infinity, the expected value of νt tends to θ.
rho = -0.2          # Rho; the correlation of the two Wiener processes. (Returns <-> Volatility in this simulation)
mu = 0.05           # mu; Expected value of Stocks := Returns
```

```python
def heston_process(ksi, kappa, theta, rho, mu, random_seed=21):
  np.random.seed(random_seed)
  dt = T/N

  mean = np.array([0, 0])
  cov = np.array([[1,rho],
                  [rho, 1]])
  Z = np.random.multivariate_normal(mean, cov, (N, sim))

  S = np.zeros(shape=(N+1, sim))
  V = np.zeros(shape=(N+1, sim))

  S[0, :] = S0
  V[0, :] = v0

  for i in range(1, N+1):
    S[i] = S[i-1] * np.exp((mu-0.5*V[i-1])*dt+np.sqrt(V[i-1]*dt)*Z[i-1, :, 0])
    V[i] = np.maximum(V[i-1] + (kappa*(theta-V[i-1])*dt + ksi*np.sqrt(V[i-1] * dt)*Z[i-1, :, 1]), 0)

  return S, V
```

```python
S, V = heston_process(ksi, kappa, theta, rho, mu)
```

```python
# Create a figure and a set of subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(S)
axs[0].set_title("Stock Price")
axs[1].plot(V)
axs[1].set_title("Volatility")

plt.show()
```

### BSM 비교

```python
# Simulation Setting
T = 1.0             # Maturity
N = 252             # Number of Time Steps
dt = T/N            # Delta time
sim = 1000         # The number of Simulations

# Initial Value
S0 = 100.0          # Inital Stock Price

# Constants of Geometry Brownian Motion
sigma = 0.3         # Volatility of Returns
mu = 0.03           # Expected Returns
```

```python
# for문 이용; 상대적 비효율성
def simulate_gbm(sigma, random_seed=21):
    np.random.seed(random_seed)
    Z = np.random.normal(size=(N, sim))
    sheet = np.zeros(shape=[N + 1, sim])
    for j in range(sim):
        sheet[0, j] = S0
        for i in range(1, N + 1):
            sheet[i, j] = sheet[i-1, j] * np.exp((mu - 0.5 * sigma ** 2)*dt + sigma*np.sqrt(dt) * Z[i-1, j])
    return sheet

start = time.time()
sheet = simulate_gbm(sigma)
print("time :", time.time() - start)
```

```python
# 행렬 연산 이용; 상대적 효율성
def simulate_gbm(sigma, random_seed=42):
  np.random.seed(random_seed)
  Z = np.random.normal(size=(N, sim))
  sheet = np.zeros(shape=[N+1, sim])
  sheet[0, :] = S0
  for i in range(1, N+1):
    sheet[i] = sheet[i-1] * np.exp((mu - 0.5 * sigma ** 2)*dt + sigma*np.sqrt(dt) * Z[i-1, :])
  return sheet

start = time.time()
sheet = simulate_gbm(sigma)
print("time :", time.time() - start)
```

```python
plt.plot(sheet)
plt.title("BSM Model Stock Price")
plt.show()
```
